## Analysing and cleaning Small Dataset

```python
import numpy as np
import pandas as pd
df = pd.read_csv("C:/Users/kkris/Desktop/Data.csv")
df
```

```python
#Approach 1: Using Imputation

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
imputer.fit(df.iloc[:,:1].values)
df.iloc[:,:1] = imputer.transform(df.iloc[:,:1].values)
df
```

```python
#Approach 2: Removing null values
df.dropna()
df
```

```python
#Approach 3: Using Imputation for numeric values

imputerr = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputerr.fit(df.iloc[:,1:3].values)
df.iloc[:,1:3] = imputerr.transform(df.iloc[:,1:3].values)
df
```

```python
imputerrr = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
imputerrr.fit(df.iloc[:,-1].values.reshape(-1, 1))
df.iloc[:,-1] = imputerrr.transform(df.iloc[:,-1].values.reshape(-1, 1))
df
```
